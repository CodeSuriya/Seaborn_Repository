```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
# In these workbook we discuss about some topics like:
# styles 
# spine removal
# size and aspect
# scale and context
```

```python
# Styles
```

```python
sns.countplot(x='sex',data=df)
```

```python
sns.set_style('white')
sns.countplot(x= 'sex',data=df)
```

```python
sns.set_style('dark')
sns.countplot(x= 'sex',data=df)
```

```python
sns.set_style('whitegrid')
sns.countplot(x= 'sex',data=df)
```

```python
sns.set_style('darkgrid')
sns.countplot(x= 'sex',data=df)
```

```python
sns.set_style('ticks')
sns.countplot(x= 'sex',data=df)
```

```python
# spine removal:
# if suppose we don't want the spine then you use the word for spine removal 'despine' and you don't want the line in left side we use inside the despine 'left = true' 
sns.set_style('white')
sns.countplot(x= 'sex',data=df)
sns.despine(left = True) 
```

```python
# Size and Aspect:
# You can use matplotlib's **plt.figure(figsize=(width,height) ** to change the size of most seaborn plots.

# You can control the size and aspect ratio of most seaborn grid plots by passing in parameters: size, and aspect. 
```

```python
# Non Grid Plot
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=df)
```

```python
# Grid Type Plot
sns.lmplot(x='total_bill',y='tip',height=2,aspect=4,data=df)
```

```python
# Scale and Context
# The set_context() allows you to override default parameters:
```

```python
sns.set_context('paper',font_scale=4)
sns.countplot(x='sex',data=df)
```

```python
sns.set_context('notebook',font_scale=4)
sns.countplot(x='sex',data=df)
```

```python
sns.set_context('talk',font_scale=4)
sns.countplot(x='sex',data=df)
```

```python
sns.set_context('poster',font_scale=4)
sns.countplot(x='sex',data=df)
```
