```python
import seaborn as sns
%matplotlib inline
```

```python
# Distribution Plots
# Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

# distplot( this distplot is update their code with one of two new functions: either displot (note, no t) or histplot.)
# jointplot
# pairplot
# rugplot
# kdeplot
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
df.head()
```

```python
# displot:
# The displot shows the distribution of a univariate set of observations.
```

```python
sns.displot(df['total_bill'])
```

```python
sns.displot(df['total_bill'],kde = True,bins = 30)
```

```python
# jointplot:
# jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what kind parameter to compare with:

# â€œscatterâ€
# â€œregâ€
# â€œresidâ€
# â€œkdeâ€
# â€œhexâ€
```

```python
sns.jointplot(x = 'total_bill', y = 'tip',data = df)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip',data = df,kind = 'reg')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip',data = df,kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip',data = df,kind = 'kde')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip',data = df, kind = 'resid')
```

```python
# pairplot:
# pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns)
# and supports a color hue argument (for categorical columns).
```

```python
sns.pairplot(df)
```

```python
# Common palettes include:

# Set1: Bright and bold colors.
# deep: Seaborn's default color palette.
# husl: A palette that ensures perceptual uniformity (colors are distinct to the human eye).
# coolwarm: A diverging color palette.
```

```python
sns.pairplot(df,hue='sex',palette = 'Set1')
```

```python
# rugplot:
# rugplots are actually a very simple concept, they just draw a dash mark for every point on a univariate distribution. 
# They are the building block of a KDE plot:
```

```python
sns.rugplot(df['total_bill'])
```

```python
# kdeplot:
# kdeplots are Kernel Density Estimation plots. These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value.
# if we want the kdeplot and rugplot in one graph use this ..

sns.kdeplot(df['total_bill'])
sns.rugplot(df['total_bill'])
```

```python
sns.kdeplot(df['tip'])
sns.rugplot(df['tip'])
```
