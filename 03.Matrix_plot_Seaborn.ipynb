```python
# Heatmap:
# In order for a heatmap to work properly, your data should already be in a matrix form, the sns.heatmap function basically just colors it in for you.
```

```python
df.head()
```

```python
# Matrix form for correlation data( these table is mixed value of numeric and non numeric .the correlation value should be calculated for 
# numerical values in the table , that's why here we using these steps to find the correlations )

numeric_data = df.select_dtypes(include=['number'])
corr_matrix = numeric_data.corr()
```

```python
corr_matrix
```

```python
sns.heatmap(corr_matrix,annot = True)
```

```python
sns.heatmap(corr_matrix,annot = True,cmap = 'coolwarm')
```
