```python
#Regression Plots
#Seaborn has many built-in capabilities for regression plots, however we won't really discuss regression until the machine learning section of the course, so we will only cover the lmplot() function for now.
#lmplot allows you to display linear models, but it also conveniently allows you to split up those plots based off of features, as well as coloring the hue based off of features.
```

```python
import seaborn as sns
%matplotlib inline
```

```python
df= sns.load_dataset('tips')
```

```python
df
```

```python
# lmplot():

sns.lmplot(x='total_bill',y= 'tip',data = df)
```

```python
sns.lmplot(x='total_bill',y= 'tip',data = df,hue= 'sex')
```

```python
sns.lmplot(x='total_bill',y= 'tip',data = df,hue= 'sex',palette = 'Set1')
```

```python
# Working with Markers:
# a scatter_kws parameter that gets passed to plt.scatter. So you want to set the s parameter in that dictionary which corresponded to the markersize
```

```python
sns.lmplot(x='total_bill',y='tip',data = df,hue='sex',palette = 'rainbow',markers=['o','v'],scatter_kws={'s':50})
```

```python
# Using a Grid
# We can add more variable separation through columns and rows with the use of a grid. Just indicate this with the col or row arguments:
```

```python
sns.lmplot(x='total_bill',y='tip',data = df,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data = df,col='time',row='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data = df,col='day',hue='sex',palette='viridis')
```

```python
# Aspect and Size
# Seaborn figures can have their size and aspect ratio adjusted with the size and aspect parameters:
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='day',hue='sex',palette='coolwarm', aspect=0.8,height=3)
```
